```python
import networkx as nx
import matplotlib.pyplot as plt
```

Edges with attributes.

```python
G = nx.Graph()
G.add_edge("A","B", weight = 6)
G.add_edge("B","C", weight = 2)
G.add_edge("C","E", weight = 3)
G.add_edge("C","F", weight = 1)
G.add_edge("D","E", weight = 8)
G.add_edge("F","G", weight = 3)
p = {"A":[-1,1],"B":[1,1],"C":[3,1],"D":[5,1],"E":[7,3],"F":[3,4],"G":[4,6]}
labels = nx.get_edge_attributes(G,'weight')
nx.draw(G,pos = p, with_labels=1)
nx.draw_networkx_edge_labels(G,pos = p,edge_labels=labels)
plt.show()
```

Nodes with attributes.

```python
G = nx.Graph()
rol = [13, 11, 13, 11, 12, 10, 12]
G.add_node("A",role = rol[0])
G.add_node("B",role = rol[1])
G.add_node("C",role = rol[2])
G.add_node("D",role = rol[3])
G.add_node("E",role = rol[4])
G.add_node("F",role = rol[5])
G.add_node("G",role = rol[6])
G.add_edge("A","B", weight = 6)
G.add_edge("B","C", weight = 2)
G.add_edge("C","E", weight = 3)
G.add_edge("C","F", weight = 1)
G.add_edge("D","E", weight = 8)
G.add_edge("F","G", weight = 3)
p = {"A":[-1,1],"B":[1,1],"C":[3,1],"D":[5,1],"E":[7,3],"F":[3,4],"G":[4,6]}
labels1 = nx.get_edge_attributes(G,'weight')
nx.draw(G, pos = p, with_labels = 1)
nx.draw_networkx_edge_labels(G,pos = p,edge_labels=labels1)
nx.draw_networkx_nodes(G, pos = p, node_color = rol)
plt.show()
```

Edges and nodes in Graph.

```python
print("Edges:",tuple(G.edges()))
print("Nodes:",tuple(G.nodes()))
```

Summary Of Graph.

```python
print(nx.info(G))
```

Edge and Node Colours.
Save graph as png file.

```python
G = nx.Graph()
wt = [1, 2, 2, 4, 3, 1]
rol = [13, 11, 13, 11, 12, 10, 12]
G.add_node("A",role = rol[0])
G.add_node("B",role = rol[1])
G.add_node("C",role = rol[2])
G.add_node("D",role = rol[3])
G.add_node("E",role = rol[4])
G.add_node("F",role = rol[5])
G.add_node("G",role = rol[6])
G.add_edge("A","B", weight  = wt[0])
G.add_edge("B","C", weight  = wt[1])
G.add_edge("C","E", weight  = wt[2])
G.add_edge("C","F", weight  = wt[3])
G.add_edge("D","E", weight  = wt[4])
G.add_edge("F","G", weight  = wt[5])
p = {"A":[-1,1],"B":[1,1],"C":[3,1],"D":[5,1],"E":[7,3],"F":[3,4],"G":[4,6]}
nx.draw(G, pos = p, with_labels = 1)
nx.draw_networkx_edges(G, pos = p, edge_color = wt)
nx.draw_networkx_nodes(G, pos = p, node_color = rol)
plt.savefig("Graph.png")
plt.show()
```

Complete Graph.

```python
H = nx.complete_graph(10)
nx.draw(H,with_labels=1)
plt.show()
```

Directed Acyclic Graph.

```python
G = nx.DiGraph()
G.add_edge("A","B")
G.add_edge("B","C")
G.add_edge("C","E")
G.add_edge("F","C")
G.add_edge("D","E")
G.add_edge("F","G")
print(nx.is_directed_acyclic_graph(G))
nx.draw(G,pos=nx.spring_layout(G),with_labels=1)
plt.show()
```

Topological sort.

```python
print(list(nx.topological_sort(G)))
```

BFS.

```python
print(list(nx.edge_bfs(G)))
```

DFS.

```python
print(list(nx.edge_dfs(G)))
```
